```python
# EXAMPLE IN SCATTER PLOT
"""Now, let's create a scatter plot and take a look at the continent's statistics, look at the number of total cases followed
by continent hover datais set to 'Country/Region' and 'Continent'."""
px.scatter(
    dataset1,
    x="Continent",
    y="TotalCases",
    hover_data=["Country/Region", "Continent"],
    color="TotalCases",
    size="TotalCases",
    size_max=80,
)
```

```python
"""log_y= True, the histogram axis (not the returned parameter) is in log scale.
The return parameter (n, bins), i.e. the values of bins and sides of bins are the
same for log=True and log=False. This means both n==n2 and bins==bins2 are true"""

px.scatter(
    dataset1,
    x="Continent",
    y="TotalCases",
    hover_data=["Country/Region", "Continent"],
    color="TotalCases",
    size="TotalCases",
    size_max=80,
    log_y=True,
)

"""By default the Y-axis is linear, but when we set log_y = True, 
then y becomes logarthmic, instad of linear.This means that instead of spacing 
numbers evenly (100, 200, .... 300), It spaces them by powers (10, 100, 1000, ... 10000)."""
```

```python
px.scatter(
    dataset1.head(60),  # THIS SHOULD ONLY USE THE FIRST 60 ROWS OF THE DATASET.
    x="Continent",
    y="TotalCases",
    size_max=90,
    hover_data=["Country/Region", "Continent"],
    color="TotalCases",
    log_y=True,  # SETTING A-AXIS TO Logarithmic, INSTEAD OF LINEAR.
    size="TotalCases",
)
```

```python
"""NOW EW CANR CREATE ANOTHER SCATTER THAT SHOWCASE ALL OF THE CONTINENT'S STATISTICS,
FIRST IT WAS BASED ON THE NUMEBR OF CASES BU NOW IT'S BASED ON THE TOTAL NUMBER OF
TESTS FOR ECH CONTINENT."""

px.scatter(
    dataset1.head(50),
    x="Continent",
    y="TotalTests",
    hover_data=["Population", "Country/Region", "Continent"],
    size_max=80,
    color="TotalTests",
    size="TotalTests",
    log_y=True,
)
```
